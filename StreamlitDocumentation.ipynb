**1. Display & Text Elements**

st.title(), st.header(), st.subheader(), st.text(), st.caption(), st.latex(), st.markdown(), st.write(), st.code()

```python
# Title - Main heading of the app
st.title("Streamlit Display & Text Elements")

# Header - Section header
st.header("1. Header Example")

# Subheader - Subsection header
st.subheader("1.1 Subheader Example")

# Text - Displaying a simple text
st.text("This is a simple text element in Streamlit.")

# Caption - Small caption text
st.caption("This is a caption text, usually used for additional information.")

# LaTeX - Displaying a mathematical formula
st.latex(r"E = mc^2")

# Markdown - Formatting text using Markdown
st.markdown("**Markdown supports bold, _italics_, and `inline code`!**")

# Write - Mixed content display (text, markdown, dataframes, etc.)
st.write("This is `st.write()`, which can handle multiple formats!")
st.write("It can display *Markdown*, **bold text**, and even numbers:", 42)

# Code - Displaying a code block
st.code('''class Developer:
    def __init__(self, name):
        self.name = name

    def show_name(self):
        return f"Developer: {self.name}"

dev = Developer("Bhuvan Ummidisetti")
print(dev.show_name())''', language="python")

```

**How to Run?**

1. Save the script as file_name.py.

2. Run the command: `streamlit run file_name.py`

3. The webpage will show all the elements in the specified order.

**Text & Numeric Inputs:**

st.text_input(),
st.text_area(),
st.number_input()


```python
# Text Input - Used to accept a single line of text input from the user
user_name = st.text_input("Enter your name:")
if user_name:
    st.write(f"Hello, {user_name}!")
# Text Area - Used to accept multi-line text input from the user
feedback = st.text_area("Provide your feedback:", height=150)
if feedback:
    st.write("Thank you for your feedback!")

# Number Input - Used to accept numeric values from the user
age = st.number_input("Enter your age:", min_value=0, max_value=100, value=25)
st.write(f"Your age is: {age}")
```

**Selection Widgets:**

st.checkbox(),
st.radio(),
st.selectbox(),
st.multiselect(),
st.slider(),
st.select_slider()

```python


# 1. Checkbox - Used to toggle between True/False states (e.g., Accepting terms, Enabling a feature)
agree = st.checkbox("I agree to the Terms & Conditions")
if agree:
    st.write("Thank you for accepting the terms!")

# 2. Radio Button - Used when only **one** option can be selected (e.g., Gender, Payment Mode)
gender = st.radio("Select your gender:", ["Male", "Female", "Other"])
st.write(f"You selected: {gender}")

# 3. Selectbox - Dropdown where the user selects **one** option (e.g., Choose a country, Select a role)
country = st.selectbox("Select your country:", ["India", "USA", "UK", "Australia"])
st.write(f"Country selected: {country}")

# 4. Multiselect - Allows selection of **multiple** options (e.g., Choosing skills, Selecting interests)
skills = st.multiselect("Select your skills:", ["Python", "Java", "C++", "JavaScript", "SQL"])
st.write(f"Selected skills: {', '.join(skills) if skills else 'None'}")

# 5. Slider - Used for selecting a **numeric range** (e.g., Selecting age, Rating a product)
age = st.slider("Select your age:", min_value=10, max_value=60, value=25)
st.write(f"Selected age: {age}")

# 6. Select Slider - Allows selection of values from a **custom range** (e.g., Mood levels, Custom categories)
mood = st.select_slider("How's your mood today?", options=["Sad", "Neutral", "Happy", "Excited"])
st.write(f"Your mood: {mood}")
```

**3. Data Display & Tables:**

st.table(),
st.dataframe(),
st.metric()


```python
import pandas as pd
# Sample Data
data = {"Name": ["Alice", "Bob", "Charlie"],
        "Age": [25, 30, 35],
        "Score": [85, 90, 95]}

df = pd.DataFrame(data)

# st.table() - Displays static table
st.subheader("Static Table")
st.table(df)

# st.dataframe() - Interactive table with sorting & filtering
st.subheader("Interactive Dataframe")
st.dataframe(df)

# st.metric() - Display a key-value pair (e.g., performance metrics)
st.subheader("Performance Metric")
st.metric(label="Stock Price", value="$150", delta="+5%")
```

**5. layouts & Containers:**

st.sidebar(),

```python
import time
# Sidebar - Creates a collapsible sidebar
st.sidebar.header("Sidebar Menu")
option = st.sidebar.selectbox("Choose an option:", ["Home", "About", "Contact"])
st.sidebar.write(f"Selected: {option}")

# Container - Grouping multiple elements together
with st.container():
    st.subheader("Container Example")
    st.write("This is inside a container.")
    st.write("It helps in grouping multiple elements together.")

# Columns - Splitting layout into columns
st.subheader("Columns Example")
col1, col2 = st.columns(2)

with col1:
    st.write("This is Column 1")

with col2:
    st.write("This is Column 2")

# Expander - Collapsible section for extra details
with st.expander("Click to expand"):
    st.write("This is a hidden section. You can put extra information here.")
```

**6. State Management & Caching:**

st.session_state(),
st.cache_data(),
st.cache_resource()

```python
st.title("State Management & Caching in Streamlit")

# 1. Session State - Store and manage user input
if "counter" not in st.session_state:
    st.session_state.counter = 0

if st.button("Increase Counter"):
    st.session_state.counter += 1

st.write(f"Counter Value: {st.session_state.counter}")

# 2. Cache Data - Caches function results (useful for data processing)
@st.cache_data
def expensive_computation(n):
    time.sleep(2)  # Simulate a slow function
    return f"Processed {n}"

if st.button("Run Expensive Computation"):
    result = expensive_computation(10)
    st.write(result)

# 3. Cache Resource - Caches resources (useful for loading models)
@st.cache_resource
def load_model():
    time.sleep(2)  # Simulate model loading time
    return "Model Loaded Successfully"

if st.button("Load Model"):
    model_status = load_model()
    st.write(model_status)
```

**8. Progress & Status Indicators:**

st.progress(),
st.toast(),

Status Message-st.error(), st.warning(), st.success(), st.info()

```python
# A progress bar is a visual representation of a task's completion status.
# It fills up as the task progresses, indicating how much work is done.
st.subheader("Progress Bar")
progress_bar = st.progress(0)  # Creates a progress bar with an initial value of 0
for percent in range(101):  # Loop to simulate progress from 0% to 100%
    time.sleep(0.02)  # Simulating a process delay
    progress_bar.progress(percent)  # Updates progress bar dynamically
st.success("Process Completed!")  # Displays a success message when progress is complete

# A toast notification is a temporary message that appears briefly to provide feedback.
# It disappears automatically after a few seconds.
st.subheader("Toast Notification")
if st.button("Show Toast Message"):  # Button to trigger the toast message
    st.toast("This is a toast notification!", icon="✅")  # Displays a floating notification

# Status messages help in communicating important information to the user.
# They can indicate errors, warnings, success, or general information.
st.subheader("Status Messages")
st.error("This is an error message!")  # Used to indicate an error (red color)
st.warning("This is a warning message!")  # Used to show a warning (yellow/orange color)
st.success("This is a success message!")  # Used to indicate a successful operation (green color)
st.info("This is an info message!")  # Used to provide general information (blue color)
```
